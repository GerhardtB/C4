## Clip Tifs

This function creates a script that clips the Tifs using Gdal.

```python
# List filepaths for all bands in the scence
FileList = glob.glob(os.path.join(r'/home/gerhardt/RadarData/C4_SDM/SSP585_2081_2100/30s','*wc2.1_30s_bioc*.tif'))

# Define clipping extent
xmin, ymin, xmax,ymax = (56.59166666666681, -20.525000000000063, 63.50000000000015, -10.341666666666725) # INSERT HERE THE CORRECT COORDINATES

# Generate gdalwarp command for each band
command = ""

for fp in FileList:
    inputfile = fp
    outputfile = inputfile[:-4] + "_clip.tif"
    
    command += "gdalwarp -te %s %s %s %s %s %s \n" % (xmin, ymin, xmax, ymax, inputfile, outputfile)
    
# Write the commands to an .sh file
cmd_file = "ClipGCM.sh"
f = open(os.path.join(cmd_file), 'w')

f.write(command)
f.close()
```

### Read GeoTifs 

This function:

1. Walks through the directory, looking for the clipped Tifs.

2. Creates image objects and appends to the list "tci_images".

3. Gets the profile of the first image (since all images are in same location).


```python
def get_imgs_and_profile(imgs_path):
    
    tci_images = []
    profile = None
    
    for root, dirs, files in os.walk(imgs_path, topdown=True):
        for fname in files:
            if fname.endswith('_clip.tif'):
                if profile is None:
                    temp_img = rio.open(os.path.join(root, fname))
                    profile = temp_img.profile
                tci_images.append(xr.open_rasterio(os.path.join(root, fname) , chunks={'x': 1000, 'y': 1000}))
    return (tci_images,profile)
```

### Get median

This function:

1. Loops over each band creating a list of relevant bands
    i.e Gets all band 1s, Concatenates band 1s and get median
2. Appends median for each band

```python
def get_median_img(imgs, no_of_bands):
    
    bands_medians = []
    for b in range(no_of_bands):
        bands = [img.sel(band=b+1) for img in imgs]
        bands_comp = xr.concat(bands, dim='band')
        bands_medians.append(bands_comp.median(dim='band', skipna=True))
    
    return xr.concat(bands_medians,dim='band')
```

# Extract Single Bands

```python
src = rio.open('/home/gerhardt/RadarData/C4_SDM/SSP585_2081_2100/Median/wc2.1_30s_bioc_MedComposite_ssp585_2081_2100.tif')

for band in range(1, src.count + 1):
    single_band = src.read(band)
    
    # get the output name
    out_name = os.path.basename('/home/gerhardt/RadarData/C4_SDM/SSP585_2081_2100/Median/wc2.1_30s_bioc_MedComposite_ssp585_2081_2100.tif')
    file, ext = os.path.splitext(out_name)
    name = file + "_" + "B" + str(band) + ".tif"
    out_img = os.path.join('/home/gerhardt/RadarData/C4_SDM/SSP585_2081_2100/Median/', name)
    print(out_img + " done")
    
    # Copy the metadata
    out_meta = src.meta.copy()
    out_meta.update({"count": 1})
    
    # save the clipped raster to disk
    with rio.open(out_img, "w", **out_meta) as dest:
        dest.write(single_band, 1)
```
