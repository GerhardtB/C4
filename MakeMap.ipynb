```python
import folium
import os
import geopandas as gpd
from pathlib import Path
from pyproj import CRS
import webbrowser
```

```python
# Read the shapefile using geopandas
shapefile_path = Path('Degradation.shp')
gdf = gpd.read_file(shapefile_path)
desired_crs = CRS.from_epsg(4326)  # Example: WGS84 CRS
gdf = gdf.to_crs(desired_crs)

# Get the centroid coordinates of the shapefile's geometry
center_lon,center_lat  = gdf['geometry'].centroid.iloc[0].coords[0]

# Create a folium map centered on the shapefile's centroid
map_obj = folium.Map(location=[center_lat, center_lon], zoom_start=8)

# Add the shapefile's geometry to the map
folium.GeoJson(gdf).add_to(map_obj)
```

```python
nia = gpd.read_file('NIA_ThreatenedEcosystems.shp')
nia = nia.to_crs(4326)
```

```python
colors = {'Highly degraded': 'red', 'Degraded': 'blue', 'Stable but stressed': 'yellow', 'Stable': 'green'}

for _, r in gdf.iterrows():
    sim_geo = gpd.GeoSeries(r["geometry"])
    geo_j = sim_geo.to_json()
    clas = r["Degradatio"]
    colr = colors.get(clas)
    geo_j = folium.GeoJson(data=geo_j, style_function=lambda x: {"fillColor": colr}, name=clas)
    folium.Popup(r["Degradatio"]).add_to(geo_j)
    geo_j.add_to(map_obj)

for _, r in nia.iterrows():
    sim_geo = gpd.GeoSeries(r["geometry"])
    geo_j = sim_geo.to_json()
    clas = r["name_2"]
    geo_j = folium.GeoJson(data=geo_j, name=clas)
    folium.Popup(r["name_2"]).add_to(geo_j)
    geo_j.add_to(map_obj)

# Create the layer control
layer_control = folium.LayerControl(show=False)

# Add the layer control to the map
layer_control.add_to(map_obj)
```
