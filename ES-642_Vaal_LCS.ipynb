```python
def head(filename,N=10):
    n=0
    with open(Path(filename)) as fp:
        for l in fp.readlines():
            if n<N:
                print(l,end="")
                n=n+1
                
def tail(filename,N=10):
    n=0
    with open(Path(filename)) as fp:
        lines=[i for i in fp.readlines()]
        Nlines=len(lines)
        for l in lines[Nlines-N:]:
            if n<N:
                print(l,end="")
                n=n+1
                
def ConvertNumeric(df,var):
    df[var]= pd.to_numeric(df[var],errors='coerce')
    
def testLogger(filename):
    Sep=","
    Data=[]
    with open(filename) as fp:
        for line in fp.readlines():
            row=line.split(Sep)
            Data.append(len(row))
    print(pd.DataFrame({"Columns":Data}).describe())
    
def ncolumnsLogger(filename):
    Sep=","
    Data=[]
    with open(filename) as fp:
        for line in fp.readlines():
            row=line.split(Sep)
            Data.append(len(row))
    return(np.median(np.array(Data)))

def readLogger(filename, nLines=18):
    DataString=""
    Sep=","
    with open(filename) as fp:
        for line in fp.readlines():
            row=line.split(Sep)
            if isinstance(nLines, list):
                if len(row) in nLines and re.match('\W*(\d{4})[/.-](\d{2})[/.-](\d{2})\s{1}(\d{2}):(\d{2}):(\d{2})\W*',row[0]):
                    DataString=DataString+line                
            else:
                if len(row) == nLines and re.match('\W*(\d{4})[/.-](\d{2})[/.-](\d{2})\s{1}(\d{2}):(\d{2}):(\d{2})\W*',row[0]):
                    DataString=DataString+line
    return(DataString)

def readRM(filename, nLines=18):
    DataString=""
    Sep=","
    with open(filename) as fp:
        for line in fp.readlines():
            row=line.split(Sep)
            if len(row) == nLines and re.match('\W*(\d{4})[/.-](\d{2})[/.-](\d{2})\s{1}(\d{2}):(\d{2}):(\d{2})\W*',row[0]):
                DataString=DataString+line
    return(DataString)

def lsR(dir,pattern=".csv",prefix="",postfix=""):
    list=[]
    for path,dirs,files in os.walk(Path(dir)):
        for f in files:
            if f.count(pattern) > 0 and f.startswith(prefix) and f.endswith(postfix):
                list.append(os.path.join(path,f))
    return(list)

def clearplots():
    plt.clf()
```

Function to read RMYoung file

```python
def readRM(filename, nLines=18, nChars=40):
    DataString=""
    Sep=","
    with open(filename,encoding='ISO-8859-1') as fp:
        for line in fp.readlines():
            if len(line)==nChars:
                row=line.split(Sep)
                if len(row) == nLines and re.match('\W*(\d{4})[/.-](\d{2})[/.-](\d{2})\s{1}(\d{2}):(\d{2}):(\d{2})',row[0]):
                    DataString=DataString+line
    return(DataString)
```

Combine files

```python
stations=["DES01","DES02","DES03","DES04","DES05","DES06","DES07","DES08"]

ESvariables=["PM2.5 (ug/m3)","Flow (l/m)","Temperature (degC)","Humidity (%)","Pressure"]
RMvariables=["Wind_Speed (m/s)","Wind_Direction (deg)"]

RawDir=Path(DataDir / 'ES642 - All Locations/RawData')
FinalDir=Path(DataDir / 'ES642 - All Locations/Data')
DailyDir=Path(DataDir / 'ES642 - All Locations/DailyData')

for s in stations:  
    dfRM=pd.DataFrame()
    dfES=pd.DataFrame()
    n=1
    files=[f for f in os.listdir(RawDir) if s in f]
    for f in files:
        print("{} of {}: ".format(n,len(files)),end="")
        n=n+1
        fullpath = os.path.join(RawDir,f)
        if f.endswith(".csv") and fullpath.count(s) > 0:
            if f.startswith("RMyoung"):
                print("Reading ",f)
                
                Data = readRM(fullpath,nLines=5)                                                                                                                                                                                                                                        
                dftmp = pd.read_csv(StringIO(Data),error_bad_lines=False,index_col=0,parse_dates=[0],names=["Date","Status","Wind_Speed (m/s)","Wind_Direction (deg)","Check"])
                if len(dfRM) == 0:
                    dfRM = dftmp
                else:
                    dfRM = pd.concat([dfRM,dftmp],axis=0)
            if f.startswith("ES642"):
                print("Reading ",f)
                Data = readRM(fullpath,nLines=8,nChars=59)
                dftmp = pd.read_csv(StringIO(Data),error_bad_lines=False,index_col=0,parse_dates=[0],names=["Date","PM2.5 (mg/m3)","Flow (l/m)","Temperature (degC)","Humidity (%)","Pressure","Status2","Checksum"])
                dftmp["PM2.5 (ug/m3)"]=dftmp["PM2.5 (mg/m3)"]*1000
                if len(dftmp) == 0:
                    dfES = dftmp
                else:
                    dfES = pd.concat([dfES,dftmp],axis=0)
                    
    print("Finished with {}".format(s))            
    if len(dfRM) > 0 and len(dfES) > 0:    
        # Make sure time ingest was succesful
        for v in ESvariables:
            ConvertNumeric(dfES,v)
        for v in RMvariables:
            ConvertNumeric(dfRM,v)

        dfRM.index=pd.to_datetime(dfRM.index, errors='coerce')
        dfES.index=pd.to_datetime(dfES.index, errors='coerce')

        # Resample datasets to 5 minute interval
        dfRM=dfRM[RMvariables].resample('5min').mean()
        dfES=dfES[ESvariables].resample('5min').mean()

        df=pd.concat([dfRM,dfES],axis=1)
        df=df['2020-07-01':]
        for idx, day in df.groupby(df.index.date):
            try:
                day.to_csv(Path(DailyDir / '{}_{}_finaldata.csv'.format(s,str(idx).replace("-",""))))
            except:
                pass
        df.to_csv(Path(FinalDir / '{}_finaldata.csv'.format(s)))
    
#    for idx, day in df.groupby(df.index.date):
#    print (day)
# Read existing file
#try:
#    df3=pd.read_csv("/home/pi/Data/{}_data.csv".format(instrument),index_col=0,parse_dates=[0]) 
#except:
#    df3=pd.DataFrame()

```
